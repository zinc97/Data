```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis

# 1. Read the CSV file
file_path = "student_interaction.csv"  
data = pd.read_csv(file_path)

# Display the first few rows of the data to verify successful loading
print("Preview of the data:")
print(data.head())
```

```python
# 2. Analyze data: Calculate total interactions and average sentiment for each student
summary = data.groupby("Student")["Interactions", "Sentiment"].agg(
    {"Interactions": "sum", "Sentiment": "mean"}
).reset_index()

# Display the summary statistics
print("\nSummary statistics:")
print(summary)

```

```python
# 3. Export data: Save the summary statistics to a new CSV file
output_path = "summary_statistics.csv"
summary.to_csv(output_path, index=False)
print(f"\nSummary data has been saved to: {output_path}")
```

```python
# Importing the necessary DataFrame class from pandas
from pandas import DataFrame

# Reading the 'Discussion forum.csv' file using pandas read_csv function
# The engine is set to 'python' to handle complex file structures
forum = pd.read_csv('Discussion forum.csv',engine='python')


# Creating an empty DataFrame with specified columns: 'Source', 'Target', and 'Weight'
# This DataFrame will be used to store the relationships between forum users
user_connect = DataFrame(columns = ['Source','Target','Weight'])
user_connect
```

```python
k=0    
for i in range(forum.shape[0]-1):    
    j = i+1
    if forum.loc[i]['author_id'] != forum.loc[j]['author_id']:
        if (forum.loc[j]['type_label']==2 and forum.loc[i]['type_label']==1) or (forum.loc[j]['type_label']==3 and forum.loc[i]['type_label']==2):
            user_connect.loc[k,'Target'] = forum.loc[i]['author_id']
            user_connect.loc[k,'Source'] = forum.loc[j]['author_id']
            user_connect.loc[k,'Weight'] = 1
            k = k+1
        if forum.loc[j]['type_label']==2 and forum.loc[i]['type_label']==2:
            kk = i
            while True:
                kk = kk - 1
                if forum.loc[kk]['type_label'] == 1:
                    break
                else:
                    continue
            user_connect.loc[k,'Target'] = forum.loc[kk]['author_id']
            user_connect.loc[k,'Source'] = forum.loc[j]['author_id']
            user_connect.loc[k,'Weight'] = 1
            k = k+1
        if forum.loc[j]['type_label']==3 and forum.loc[i]['type_label']==3:
            kk = i
            while True:
                kk = kk - 1
                if forum.loc[kk]['type_label'] == 2:
                    break
                else:
                    continue
            user_connect.loc[k,'Target'] = forum.loc[kk]['author_id']
            user_connect.loc[k,'Source'] = forum.loc[j]['author_id']
            user_connect.loc[k,'Weight'] = 1
            k = k+1 
        if forum.loc[j]['type_label']==2 and forum.loc[i]['type_label']==3:
            kk = i
            while True:
                kk = kk - 1
                if forum.loc[kk]['type_label'] == 1:
                    break
                else:
                    continue
            user_connect.loc[k,'Target'] = forum.loc[kk]['author_id']
            user_connect.loc[k,'Source'] = forum.loc[j]['author_id']
            user_connect.loc[k,'Weight'] = 1
            k = k+1              
    else:
        continue
    print('第 %d 条帖子'%(i))
#user_connect['Target'] = user_connect['Target'].astype(int)
#user_connect['Source'] = user_connect['Source'].astype(int)
```

```python
# Importing the networkx library for graph creation and visualization
import networkx as nx

# Create a directed graph (DiGraph) object to store weighted edges
G_weighted = nx.DiGraph()

for i in range(0, len(user_connect)):
    G_weighted.add_edge(user_connect.loc[i]['Source'], 
                        user_connect.loc[i]['Target'], 
                        weight=1)  

# Generate the layout for the network graph. Here, spring_layout arranges nodes using a force-directed algorithm.
nx.spring_layout(G_weighted)

# Draw the network graph with the edges colored blue ('b')
nx.draw_networkx(G_weighted, edge_color='b') 
```

```python
# Network level
print(nx.density(G_weighted))
print(nx.reciprocity(G_weighted))

# Node level
print(nx.betweenness_centrality(G_weighted))
```
